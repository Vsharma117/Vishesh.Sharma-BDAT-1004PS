```python
import pandas as pd

#Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep="|")

#Mean age per occupation
mean_age_by_occupation = users.groupby("occupation")["age"].mean()
print(mean_age_by_occupation)

#Male ratio per occupation 
male_count = users[users["gender"] == "M"].groupby("occupation").size()
total_count = users.groupby("occupation").size()
male_ratio_by_occupation = (male_count / total_count) * 100
male_ratio_by_occupation = male_ratio_by_occupation.sort_values(ascending=False)
print(male_ratio_by_occupation)

#Minimum and maximum ages per occupation
age_stats_by_occupation = users.groupby("occupation")[["age"]].agg(["min", "max"])
print(age_stats_by_occupation)

#Mean age per occupation and gender
mean_age_by_occupation_gender = users.groupby(["occupation", "gender"])["age"].mean()
print(mean_age_by_occupation_gender)

#Percentage of women and men per occupation
occupations = users["occupation"].unique()

for occupation in occupations:
    occupation_data = users[users["occupation"] == occupation]
    female_count = (occupation_data["gender"] == "F").sum()
    male_count = (occupation_data["gender"] == "M").sum()
    total_count = occupation_data.shape[0]

    female_percentage = (female_count / total_count) * 100
    male_percentage = (male_count / total_count) * 100

    print(f"\nOccupation: {occupation}")
    print(f"  - Female Percentage: {female_percentage:.2f}%")
    print(f"  - Male Percentage: {male_percentage:.2f}%")
```

```python
import pandas as pd

#Step 2 Import the dataset (assuming an internet connection)
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

#Step 4 Select only the Goal column
goals = euro12["Goals"]

#Step 5 Count participating teams
num_teams = euro12["Team"].nunique()
print(f"Number of teams participated in Euro 2012: {num_teams}")

#Step 6 Get number of columns (alternative approach)
num_columns = len(euro12.columns)
print(f"Number of columns in the dataset: {num_columns}")

#Step 10 Filter teams with more than 6 goals
high_scoring_teams = euro12[euro12["Goals"] > 6]
print("\nTeams that scored more than 6 goals:")
print(high_scoring_teams)

#Step 11 Filter teams starting with "G"
teams_with_g = euro12[euro12["Team"].str.startswith("G")]
print("\nTeams that start with 'G':")
print(teams_with_g)

#Step 12 Select the first 7 columns
first_seven = euro12.iloc[:, :7]
print("\nFirst 7 columns:")
print(first_seven)

#Step 13 Select all columns except the last 3
all_except_last_three = euro12.iloc[:, :-3]
print("\nAll columns except the last 3:")
print(all_except_last_three)

#Step 14 Shooting Accuracy for England, Italy, Russia
shooting_accuracy = {
    "England": euro12[euro12["Team"] == "England"]["Shooting Accuracy"].values[0],
    "Italy": euro12[euro12["Team"] == "Italy"]["Shooting Accuracy"].values[0],
    "Russia": euro12[euro12["Team"] == "Russia"]["Shooting Accuracy"].values[0],
}
print("\nShooting Accuracy:")
print(shooting_accuracy)

#Step 7 Discipline DataFrame
discipline = euro12[["Team", "Yellow Cards", "Red Cards"]]

#Step 8 Sort by Red Cards, then Yellow Cards
disciplined_teams = discipline.sort_values(by=["Red Cards", "Yellow Cards"])
print("\nTeams sorted by Red Cards (ascending) then Yellow Cards (ascending):")
print(disciplined_teams)

#Step 9 Mean Yellow Cards
avg_yellow_cards = discipline["Yellow Cards"].mean()
print(f"\nMean Yellow Cards per Team: {avg_yellow_cards:.2f}")
```

```python
import pandas as pd

#Step 2 Import the dataset
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')

#Step 4 See the first 10 entries
print(chipo.head(10))

#Step 5 Number of observations
print("Number of observations:", len(chipo))

#Step 6 Number of columns
print("Number of columns:", len(chipo.columns))

#Step 7 Column names
print("Column names:", chipo.columns)

#Step 8 Index
print("Index:", chipo.index)

#Step 9 Most-ordered item
print("Most-ordered item:", chipo['item_name'].value_counts().idxmax())

#Step 10 Quantity of most-ordered item
print("Quantity of most-ordered item:", chipo['item_name'].value_counts().max())

#Step 11 Most ordered item in choice_description
print("Most ordered item in choice_description:", chipo['choice_description'].value_counts().idxmax())

#Step 12 Total items ordered
print("Total items ordered:", chipo['quantity'].sum())

#Step 13 Convert item price to float
chipo['item_price'] = chipo['item_price'].astype(str).str.replace('$', '', regex=False)
chipo['item_price'] = chipo['item_price'].astype(float)
print("Item price type:", chipo['item_price'].dtype)

#Step 14 Total revenue
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Total revenue:", revenue)

#Step 15 Total orders
print("Total orders:", len(chipo))

#Step 16 Average revenue per order
avg_revenue_per_order = revenue / len(chipo)
print("Average revenue per order:", avg_revenue_per_order)

#Step 17 Number of different items
print("Number of different items:", len(chipo['item_name'].unique()))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Read the data from the CSV file
df = pd.read_csv("roman-emperor-reigns.csv")

#Define the keywords indicating assassination
assassination_keywords = ['Assassinated', 'Possibly assassinated']

#Filter the data for assassinated emperors
assassinated_emperors = df[df['Cause_of_Death'].str.contains('|'.join(assassination_keywords), case=False)].count()['Emperor']

#Count the total number of emperors
total_emperors = df.shape[0]

#Calculate the fraction of assassinated emperors
fraction_assassinated = assassinated_emperors / total_emperors

#Calculate the fraction of emperors who died of natural causes
fraction_natural_causes = 1 - fraction_assassinated

#Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie([fraction_assassinated, fraction_natural_causes], labels=['Assassinated', 'Other Causes'], autopct='%1.1f%%', startangle=140)
plt.title('Fraction of Roman Emperors Assassinated')

#Equal aspect ratio ensures a circular pie chart
plt.axis('equal')

#Show the plot
plt.show()
```
